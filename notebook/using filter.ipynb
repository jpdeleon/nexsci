```python
from nexsci import downloader as dl

keyword = 'all'
df=dl.download_nexsci(keyword,replace=False)
```

```python
#select spectral type column, then get the right element after space
sptype=df['st_spstr'].apply(lambda x: str(x).split(' ')[-1])

#select the one with III or IV in that column
mask = []
for i in sptype:
    if i == 'III':
        mask.append(True)
    elif i == 'IV':
        mask.append(True)
    else:
        mask.append(False)
```

```python
#remove first column
df=df.loc[mask].drop('Unnamed: 0',axis=1)
#save as csv
df.to_csv('SpTypeIII-IV.csv',index=False)
```
